```python
import torch

```

```python

def f(x):
  return x**2/3 + torch.sin(torch.pi*x)

y_true = 1
```

```python
start_point = torch.tensor(1.2, device='cuda', requires_grad=True)
start_point
```

```python
optimazer = torch.optim.SGD(
    [start_point],
    lr= 1e-3
)
```

```python
for i in range(2000):
    res_func = f(start_point)
    error = (y_true - res_func) ** 2
    error.backward()
    optimazer.step()
    optimazer.zero_grad()

    if i % 100 == 0:
        print(error.item())

print()
print(start_point.item())
print(f(start_point).item())
```
